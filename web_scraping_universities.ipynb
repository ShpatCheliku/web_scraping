```python
import requests
from bs4 import BeautifulSoup
```

```python
website_url = requests.get('https://thegreenestworkforce.ca/index.php/en/schools/').text
```

```python
website_url
```

```python
soup = BeautifulSoup(website_url,"html.parser")
```

```python
print(soup.prettify())
```

```python
My_table = soup.find('article',{'id':'post-17896'})
My_table
```

```python
links = My_table.findAll('a')
links
```

```python
colleges = []
for link in links:
    colleges.append(link.get('title'))
    
print(colleges)
```

```python
college_links = []
for link in links:
    college_links.append(link.get('href'))
    
print(colleges)
```

```python
import pandas as pd
df = pd.DataFrame()
df['University'] = colleges
df['Website'] = college_links
df
```

```python
df.to_excel("universities_canada.xlsx",
             sheet_name='Universities',index = False)  
```
